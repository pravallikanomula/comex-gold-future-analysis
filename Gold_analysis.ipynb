# Introduction
Gold is not just a precious metal; it also serves as a key economic barometer. Its price movements reflect investor sentiment, global economic stability, inflation expectations, and geopolitical tensions, making it a critical asset for both portfolio diversification and macroeconomic analysis. Even seasoned market analysts often find its behavior difficult to predict due to the interplay of multiple economic, financial, and geopolitical factors.

# Data Loading

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("GC_in_1_hour_new.csv")
```

```python
df.head()
```

# Data Cleaning & Processing

```python
df['datetime'] = pd.to_datetime(df['datetime'])
df.set_index('datetime', inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df['volume'].replace(0, np.nan, inplace = True)
df['volume'].fillna(df['volume'].rolling(5).mean(), inplace = True)
```

```python
df['year'] = df.index.year
df['month'] = df.index.month
df['day'] = df.index.day
df['weekday'] = df.index.weekday
```

# Data Analysis

```python
plt.figure(figsize=(15,5))
plt.plot(df['close'], label='Close Price')
plt.title('Gold Daily Close Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
df['MA50'] = df['close'].rolling(50).mean()
df['MA200'] = df['close'].rolling(200).mean()
df['MA500'] = df['close'].rolling(500).mean()

plt.figure(figsize = (15,7))
plt.plot(df['close'], label = 'Close', alpha = 0.7)
plt.plot(df['MA50'], label = 'MA50', alpha = 0.5)
plt.plot(df['MA200'], label = 'MA200', alpha = 0.5)
plt.plot(df['MA500'], label = 'MA500', alpha = 0.5)
plt.legend()
plt.show()
```

```python
#Outlier detection: Boxplots over years or rolling windows

df['return'] = df['close'].pct_change()
pivot = df.pivot_table(values='return', index='year', columns='month', aggfunc='mean')
sns.heatmap(pivot, cmap='RdYlGn', center=0)
plt.title('Average Monthly Returns Heatmap')
plt.show()
```
